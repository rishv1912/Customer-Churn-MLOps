```python
features = ["account_length","area_code", "voice_mail_plan", "number_vmail_messages", "total_day_minutes", "total_day_calls", 
"total_day_charge", "total_eve_minutes", "total_eve_calls", "total_eve_charge", "total_night_minutes", "total_night_calls", "total_night_charge", 
"total_intl_minutes", "total_intl_calls", "total_intl_charge","number_customer_service_calls"]

fig, ax = plt.subplots(nrows=6,ncols=3, figsize=(18,20))
ax = ax.flat

for i,feature in enumerate(features):
    ax[i].hist(df[feature],bins=15,color='skyblue',edgecolor='black')
    ax[i].set_title(feature)
    ax[i].tick_params(axis='x',rotation=45)

for j in range(len(features),len(ax)):
    fig.delaxes(ax[j])

fig.tight_layout()
plt.show()

fig.savefig("figure.png")
```

```python
def plot_histogram(df,column_name):

    plt.figure(figsize=(8,5))
    sns.histplot(df[column_name],kde=True)
    plt.title(f"Distribution of {column_name}")

    col_mean = df[column_name].mean()
    col_median = df[column_name].median()

    plt.axvline(col_mean,color='red',linestyle='--',label='Mean')
    plt.axvline(col_median, color='green',linestyle='-',label="Median")

    plt.legend()
    plt.show()
```

```python
plot_histogram(df,"total_day_minutes")
```
